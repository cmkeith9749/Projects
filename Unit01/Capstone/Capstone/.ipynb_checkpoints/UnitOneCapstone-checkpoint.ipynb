```python
# using zillow data for 5 metro atlanta counties
regions = ['Fulton County','Gwinnett County','Dekalb County','Cobb County','Cherokee County']
labels =  [region.split()[0].lower() for region in regions]

# process zillow one variable csv files into dfs for plotting
# parameter dicts are zillow file attributes
def zillows_dfs(dict):
    cols = [2] + [i for i in range(5,dict['shape'][1])]
    
    df = pd.read_csv(dict['file']) 
    df = df[(df.StateName == 'GA') & (df.RegionName.isin(regions))]
    df = df[df.columns[cols]]
    df = df.T[1:]
    df.columns = labels
    df = df.astype(float)
    
    #df's used for plots
    df['avg'] = np.around(df.mean(axis=1), decimals=2)
    df['mth'] = [pd.Timestamp(x).month for x in df.index]
    df['doy'] = [pd.Timestamp(x).dayofyear + 14 for x in df.index]
    #df['dte'] = df. index
    df.index = [pd.Timestamp(x) for x in df.index]
    dict['df'] = df
    
    # df1's used for creating example days on market population
    df1 = df[:-3].groupby('mth').mean()[['avg', 'doy']]
    df1['jan'] = np.around(df1.avg.div(df1.avg[1]),decimals=2)
    df1.doy = df1.doy.astype(int)
    dict['table'] = df1
    
    dict['xlabel'] = 'Months'
    
    return dict
```

```python
doz = zillows_dfs(doz)
stl = zillows_dfs(stl)
mls = zillows_dfs(mls)
```
